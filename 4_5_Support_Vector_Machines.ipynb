```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVC
```

```python
X = np.array([
              [-1,-1],
              [-2,-1],
              [1,1],
              [2,1] 
])
```

```python
y = np.array([1,1,2,2])
```

```python
clf = make_pipeline(StandardScaler(), SVC(gamma='auto'))
```

```python
clf.fit(X,y)
```

```python
clf.predict([[2,-1]])
```
